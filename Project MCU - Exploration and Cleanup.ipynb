```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests

api_key = "efd18417"
```

```python
# Create base url
base_url = f"http://www.omdbapi.com/?apikey={api_key}&t="

# Create list of MCU movies
movies = ["Iron Man",
         "The Incredible Hulk",
         "Iron Man 2",
         "Thor",
         "Captain America: The First Avenger",
         "The Avengers",
         "Iron Man 3",
         "Thor: The Dark World",
         "Captain America: The Winter Soldier",
         "Guardians of the Galaxy",
         "Avengers: Age of Ultron",
         "Ant-Man",
         "Captain America: Civil War",
         "Doctor Strange",
         "Guardians of the Galaxy Vol. 2",
         "Spider-Man: Homecoming",
         "Thor: Ragnarok",
         "Black Panther",
         "Avengers: Infinity War",
         "Ant-Man and the Wasp",
         "Captain Marvel",
         "Avengers: Endgame",
         "Spider-Man: Far From Home"]
```

```python
# create list of responses
responses = []

# create loop to extract data using movies list
for movie in movies:
    movie_data = requests.get(base_url + movie).json()
    responses.append(movie_data)
```

```python
# create and show df
omdb_df = pd.DataFrame(responses)
omdb_df.head()
```

```python
#Convert to CSV
omdb_df.to_csv("output_data/omdb.csv")
```

```python
#Fetch that Box Office data 
mojofile = pd.read_csv("databases/boxofficemojodata.csv")

#Create DataFrame
bomojo = pd.DataFrame(mojofile)
bomojo.head()
```

```python
#Fetch that Opening Weekend Data
wkndfile = pd.read_csv("databases/opening_weekend_data.csv", encoding="utf-8")

#Create DataFrame
weeknd = pd.DataFrame(wkndfile)
weeknd.head()
```

```python
#insert another big title here - Merge and Cleanup Our Data
```

```python
#Merge these two
merge_df = pd.merge(weeknd, bomojo, left_on="Film", right_on="title", how="inner")

#Show the merge
merge_df
```

```python
#Now we merge the data from merge_df with the converted API. (left join)
join_df = pd.merge(omdb_df, merge_df, left_on='imdbID', right_on='imdb_title_identifier', how='left')

join_df.head()
```

```python
#Let's see those columns
join_df.columns
```

```python
#Clean the DataFrame
cleanjoin_df=join_df[['Title', 'imdbID', 'Year_x', 'Released', 'Metascore', 'imdbRating', 'imdbVotes', 'Budget', 'Opening ', 'domestic_gross', 'international_gross', 'worldwide_gross', 'widest_release', 'Production',
                      'Distributor', 'Director', 'Awards']]


#Show DataFrame
cleanjoin_df.head()
```

```python
#Bring UMC dataset from other file
umcfile = pd.read_csv("databases/umc-database.csv")

#Create UMC DataFrame
umc = pd.DataFrame(umcfile)
umc.head()
```

```python
#Merge once more!!

endgame_df = pd.merge(umc, cleanjoin_df, left_on='title', right_on='Title',how='outer')

#Show DataFrame
endgame_df.head()
```

```python
#Read Columns to see what we want to organize
endgame_df.columns
```

```python
#Clean final_df
final_df = endgame_df[['Title', 'imdbID', 'Year_x',
       'Released', 'movie_score', 'Metascore', 'imdbRating', 'imdbVotes', 'Budget',
       'Opening ', 'domestic_gross', 'international_gross', 'worldwide_gross',
       'widest_release', 'Production', 'Distributor', 'Director', 'Awards']]
final_df.head()
```

```python
#RENAME TIME
final_df = final_df.rename(columns= {'imdbID':'IMDbID', 'Year_x':'Year', 'movie_score':'Tomatometer', 'imdbRating':'IMDb', 'imdbVotes':'# of IMDb Votes', 
       'Opening ': 'Opening Weekend Sales', 'domestic_gross':'Domestic Gross', 'international_gross':'International Gross', 'worldwide_gross':'Worldwide Gross',
       'widest_release':'Widest Theater Release', 'Production':'Production House'})

#Set the Index to the Identifier
final_df=final_df.set_index('IMDbID')


final_df.head()
```

```python
#Convert the Financial Amounts to a numeric Format
final_df["Budget"] = pd.to_numeric(final_df["Budget"])
final_df["Opening Weekend Sales"] = pd.to_numeric(final_df["Opening Weekend Sales"])
final_df["Domestic Gross"] = pd.to_numeric(final_df["Domestic Gross"])
final_df["International Gross"] = pd.to_numeric(final_df["International Gross"])
final_df["Worldwide Gross"] = pd.to_numeric(final_df["Worldwide Gross"])
final_df["Metascore"] = pd.to_numeric(final_df["Metascore"])
final_df["Tomatometer"] = pd.to_numeric(final_df["Tomatometer"])
final_df["IMDb"] = pd.to_numeric(final_df["IMDb"])
final_df["Widest Theater Release"] = pd.to_numeric(final_df["Widest Theater Release"])

final_df.head()
```

```python
#Export into a CSV
final_df.to_csv('output_data/final_df.csv', index = True)

#Show DF and all film titles please
final_df.head(25)
```
