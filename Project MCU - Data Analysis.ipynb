```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
```

```python
#Import Final_df.csv

final_df = pd.read_csv('output_data/final_df.csv')

#Show
final_df.head(25)
```

```python
#Total Box Office Revenue by Release Date (Line Chart)

# Get data for our "Release Date" and "Worldwide Gross"
x_axis = final_df['Released']
y_axis = final_df['Worldwide Gross']
# Show the line plot
plt.plot(x_axis, y_axis)
plt.title("Total Box Office Revenue by Release Date ", fontsize=18)
plt.xlabel("Release Date")
plt.ylabel("Worldwide Gross (In USD Billions)")
plt.xticks(rotation='90')
plt.grid()

#Export as PNG
plt.savefig("output_data/box_office_by_release.png")

#Show Line Graph
plt.show()
```

```python
#Critic Ratings (Bar Chart)

#Create a bar chart for Rotten Tomatoes and Metacritic
x_axis = final_df['Title']
rotten_tomato = final_df["Tomatometer"]
meta_score = final_df["Metascore"]
imdb = final_df["IMDb"]

#Create Bar Chart
plt.bar(x_axis, rotten_tomato, color='r', alpha=0.5, align="center", label="Rotten Tomatoes")
plt.bar(x_axis, meta_score, color='b', alpha=0.5, align="center", label="Metascore")

#Labels and Formatting
plt.xticks(rotation='90')
plt.title("Critic Ratings (Source: Rotten Tomatoes and Metacritic)")
plt.xlabel("Movie Title")
plt.ylabel("Critic Score")
plt.legend(loc="best")
plt.grid(axis="y")
plt.ylim(0,125)

#Export as PNG
plt.savefig("output_data/RT_and_MC.png")

#Show Figure
plt.show()
```

```python
#Create Bar Chart for IMDb
plt.bar(x_axis, imdb, color='y', alpha=0.5, align="center", label="IMDb")
plt.xticks(rotation='90')

#Labels 
plt.xlabel("Movie Title")
plt.ylabel("IMDb Score")
plt.title("IMDb Scores for Each Film")
plt.grid(axis="y")
plt.ylim(0,10)

#Export as PNG
plt.savefig("output_data/IMDbScores.png")

#Show our Bar Graph
plt.show()
```

```python
#Measure Profitability (Budget vs Worldwide)

#Do A side-by-side-by-side

profitability = final_df["Worldwide Gross"]- final_df["Budget"]
mr_worldwide = final_df["Worldwide Gross"]
budget = final_df["Budget"]

plt.bar(x_axis, mr_worldwide, color='g', alpha=0.7, align="center", label="WorldWide Gross")
plt.bar(x_axis, budget, color='r', alpha=1, align="center", label="Overall Budget")
#plt.bar(x_axis, profitability, color='g', alpha=1, align="center")
plt.xticks(rotation='90')

#Go back to put Labels
plt.title("Worldwide Gross vs. Budget")


#Legend
plt.legend(loc="best")
plt.ylabel("Worldwide Gross in USD Billions")
plt.xlabel("Movie Title")
plt.grid(axis="y")


#Export as PNG
plt.savefig("output_data/WorldwideVBudget.png")

#Show our Bar Graph
plt.show()
```

```python
#Opening Box Office Numbers (Chronologically)

#Create Variable
opening = final_df["Opening Weekend Sales"]

#Create Line Plot
plt.plot(x_axis, opening, color='g')

#Create Labels and Formatting
plt.xticks(rotation='90')
plt.title("Opening Box Office Gross in Chronological Order")
plt.ylabel("Opening Weekend Sales(In USD Hundred Millions)")
plt.xlabel("Movie Title")
plt.grid()

#Export as PNG
plt.savefig("output_data/OpeningBoxOffice.png")

#Show Plot
plt.show()
```

```python
#Domestic vs International Trend
domestic = final_df["Domestic Gross"]
international = final_df["International Gross"]

#Create Line Plot
plt.plot(x_axis, domestic, color='b', label="Domestic Gross")
plt.plot(x_axis, international, color='r', label = "International Gross")

#Create Label
plt.xticks(rotation='90')
plt.xlabel("Movie Title")
plt.title("Domestic Gross vs. International Gross")
plt.ylabel("Gross by USD Billions")

#Create Legend
plt.legend(loc="best")
plt.grid()
#Export as PNG
plt.savefig("output_data/DomesticVInternational.png")

#Show Plot
plt.show()
```

```python
#Critic Vs Box Office (Scatter Plot)
from scipy.stats import linregress
import scipy.stats as sts

#Reiterate Variables
rotten_tomato = final_df["Tomatometer"]
mr_worldwide = final_df["Worldwide Gross"]

#Perform Linear Regression
(slope, intercept, rvalue, pvalue, stderr) = linregress(rotten_tomato, mr_worldwide)
regress_values = rotten_tomato * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

#Correlation Calculation (Nick came up with adding this)
correlation = sts.pearsonr(rotten_tomato, mr_worldwide)

#Create Scatter Plot
plt.scatter(rotten_tomato, mr_worldwide, facecolor='r',edgecolors="g", s=rotten_tomato)
plt.plot(rotten_tomato, regress_values, "b-")
plt.annotate(line_eq,(70,2500000000),fontsize=15,color="red")

#Create Labels
plt.title("Critical Reception (Tomatometer) vs Worldwide Box Office")
plt.xlabel("Tomatomater Score (Source: Rotten Tomatoes)")
plt.ylabel("Worldwide Box Office by USD Billions")
plt.grid()


#Export as PNG
plt.savefig("output_data/CriticalReceptionVWorldwide.png")


#Print Correlation
print(f'The correlation coefficient between our Rotten Tomatoes scores and the Worldwide Gross is {correlation[0]}')
plt.show()
```
