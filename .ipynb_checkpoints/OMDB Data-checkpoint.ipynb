```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests

api_key = "efd18417"
```

```python
# Create base url
base_url = f"http://www.omdbapi.com/?apikey={api_key}&t="

# Create list of MCU movies
movies = ["Iron Man",
         "The Incredible Hulk",
         "Iron Man 2",
         "Thor",
         "Captain America: The First Avenger",
         "The Avengers",
         "Iron Man 3",
         "Thor: The Dark World",
         "Captain America: The Winter Soldier",
         "Guardians of the Galaxy",
         "Avengers: Age of Ultron",
         "Ant-Man",
         "Captain America: Civil War",
         "Doctor Strange",
         "Guardians of the Galaxy Vol. 2",
         "Spider-Man: Homecoming",
         "Thor: Ragnarok",
         "Black Panther",
         "Avengers: Infinity War",
         "Ant-Man and the Wasp",
         "Captain Marvel",
         "Avengers: Endgame",
         "Spider-Man: Far From Home"]
```

```python
# create list of responses
responses = []

# create loop to extract data using movies list
for movie in movies:
    movie_data = requests.get(base_url + movie).json()
    responses.append(movie_data)
```

```python
# create and show df
omdb_df = pd.DataFrame(responses)
omdb_df.head()
```

```python
#Convert to CSV
omdb_df.to_csv("output_data/omdb.csv")
```

```python
#Fetch that Box Office data 
mojofile = pd.read_csv("databases/boxofficemojodata.csv")

#Create DataFrame
bomojo = pd.DataFrame(mojofile)
bomojo.head()
```

```python
#Fetch that Opening Weekend Data
wkndfile = pd.read_csv("databases/opening_weekend_data.csv", encoding="utf-8")

#Create DataFrame
weeknd = pd.DataFrame(wkndfile)
weeknd.head()
```

## Merge and Cleanup Our Data

```python
#Merge these two
merge_df = pd.merge(weeknd, bomojo, left_on="Film", right_on="title", how="inner")

#Show the merge
merge_df
```

```python
#Now we merge the data from merge_df with the converted API. (left join)
join_df = pd.merge(omdb_df, merge_df, left_on='imdbID', right_on='imdb_title_identifier', how='left')

join_df.head()
```

```python
#Let's see those columns
join_df.columns
```

```python
#Clean the DataFrame
cleanjoin_df=join_df[['Title', 'imdbID', 'Year_x', 'Released', 'Metascore', 'imdbRating', 'imdbVotes', 'Budget', 'Opening ', 'domestic_gross', 'international_gross', 'worldwide_gross', 'widest_release', 'Production',
                      'Distributor', 'Director', 'Awards']]


#Show DataFrame
cleanjoin_df.head()
```

```python
#Bring UMC dataset
umcfile = pd.read_csv("databases/umc-database.csv")

#Create UMC DataFrame
umc = pd.DataFrame(umcfile)
umc.head(0)
```

```python
#Merge once more!!

endgame_df = pd.merge(umc, cleanjoin_df, left_on='title', right_on='Title',how='outer')

#Show DataFrame
endgame_df.head()
```

```python
#Read Columns to see what we want to organize
endgame_df.columns
```

```python
#Clean final_df
final_df = endgame_df[['Title', 'imdbID', 'Year_x',
       'Released', 'movie_score', 'Metascore', 'imdbRating', 'imdbVotes', 'Budget',
       'Opening ', 'domestic_gross', 'international_gross', 'worldwide_gross',
       'widest_release', 'Production', 'Distributor', 'Director', 'Awards']]
final_df.head()
```

```python
#RENAME TIME
final_df = final_df.rename(columns= {'imdbID':'IMDbID', 'Year_x':'Year', 'movie_score':'Tomatometer', 'imdbRating':'IMDb', 'imdbVotes':'# of IMDb Votes', 
       'Opening ': 'Opening Weekend Sales', 'domestic_gross':'Domestic Gross', 'international_gross':'International Gross', 'worldwide_gross':'Worldwide Gross',
       'widest_release':'Widest Theater Release', 'Production':'Production House'})

#Set the Index to the Identifier
final_df=final_df.set_index('IMDbID')


final_df.head()
```

```python
#Convert the Financial Amounts to a Dollar Format
final_df["Budget"] = pd.to_numeric(final_df["Budget"])
final_df["Opening Weekend Sales"] = pd.to_numeric(final_df["Opening Weekend Sales"])
final_df["Domestic Gross"] = pd.to_numeric(final_df["Domestic Gross"])
final_df["International Gross"] = pd.to_numeric(final_df["International Gross"])
final_df["Worldwide Gross"] = pd.to_numeric(final_df["Worldwide Gross"])
final_df["Metascore"] = pd.to_numeric(final_df["Metascore"])
final_df["Tomatometer"] = pd.to_numeric(final_df["Tomatometer"])
final_df["IMDb"] = pd.to_numeric(final_df["IMDb"])
final_df["Widest Theater Release"] = pd.to_numeric(final_df["Widest Theater Release"])


final_df.head(25)
```

# Charts for Our Data

```python
#Total Box Office Revenue by Release Date (Line Chart)

# Get data for m601
x_axis = final_df['Released']
y_axis = final_df['Worldwide Gross']
# Show the line plot
plt.plot(x_axis, y_axis)
plt.title("Total Box Office Revenue by Release Date ", fontsize=18)
plt.xlabel("Release Date")
plt.ylabel("Worldwide Gross (In USD Billions)")
plt.xticks(rotation='90')

#plt.xlim(0,45)
#plt.ylim(25, 47)
#plt.tight_layout()
plt.show()
```

```python
#Critic Ratings (Bar Chart)

#Create a bar chart for Rotten Tomatoes and Metacritic
x_axis = final_df['Title']
rotten_tomato = final_df["Tomatometer"]
meta_score = final_df["Metascore"]
imdb = final_df["IMDb"]

#Create Bar Chart
plt.bar(x_axis, rotten_tomato, color='r', alpha=0.5, align="center", label="Rotten Tomatoes")
plt.bar(x_axis, meta_score, color='b', alpha=0.5, align="center", label="Metascore")

#Labels and Formatting
plt.xticks(rotation='90')
plt.title("Critic Ratings (Source: Rotten Tomatoes and Metacritic)")
plt.xlabel("Movie Title")
plt.ylabel("Critic Score")
plt.legend(loc="best")
plt.show()
```

```python
#Create Bar Chart for IMDb
plt.bar(x_axis, imdb, color='y', alpha=0.5, align="center", label="IMDb")
plt.xticks(rotation='90')

#Labels 
plt.xlabel("Movie Title")
plt.ylabel("IMDb Score")
plt.title("IMDb Scores for Each Film")
plt.show()
```

```python
#Measure Profitability (Budget vs Worldwide)

#Do A side-by-side-by-side

profitability = final_df["Worldwide Gross"]- final_df["Budget"]
mr_worldwide = final_df["Worldwide Gross"]
budget = final_df["Budget"]

plt.bar(x_axis, mr_worldwide, color='g', alpha=0.7, align="center", label="WorldWide Gross")
plt.bar(x_axis, budget, color='r', alpha=1, align="center", label="Overall Budget")
#plt.bar(x_axis, profitability, color='g', alpha=1, align="center")
plt.xticks(rotation='90')

#Go back to put Labels
plt.title("Worldwide Gross vs. Budget")


#Legend
plt.legend(loc="best")
plt.ylabel("Worldwide Gross in USD Billions")
plt.xlabel("Movie Title")

plt.show()
```

```python
#Opening Box Office Numbers (Chronologically)

#Create Variable
opening = final_df["Opening Weekend Sales"]

#Create Line Plot
plt.plot(x_axis, opening, color='g')

#Create Labels and Formatting
plt.xticks(rotation='90')
plt.title("Opening Box Office Gross in Chronological Order")
plt.ylabel("Opening Weekend Sales(In USD Hundred Millions)")
plt.xlabel("Movie Title")


#Show Plot
plt.show()
```

```python
#Domestic vs International Trend
domestic = final_df["Domestic Gross"]
international = final_df["International Gross"]

#Create Line Plot
plt.plot(x_axis, domestic, color='b', label="Domestic Gross")
plt.plot(x_axis, international, color='r', label = "International Gross")

#Create Label
plt.xticks(rotation='90')
plt.xlabel("Movie Title")
plt.title("Domestic Gross vs. International Gross")
plt.ylabel("Gross by USD Billions")

#Create Legend
plt.legend(loc="best")

#Show Plot
plt.show()
```

```python
#Critic Vs Box Office (Scatter Plot)
from scipy.stats import linregress
import scipy.stats as sts

#Reiterate Variables
rotten_tomato = final_df["Tomatometer"]
mr_worldwide = final_df["Worldwide Gross"]

#Perform Linear Regression
(slope, intercept, rvalue, pvalue, stderr) = linregress(rotten_tomato, mr_worldwide)
regress_values = rotten_tomato * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

#Correlation Calculation (Nick came up with adding this)
correlation = sts.pearsonr(rotten_tomato, mr_worldwide)

#Create Scatter Plot
plt.scatter(rotten_tomato, mr_worldwide, facecolor='r',edgecolors="g", s=rotten_tomato)
plt.plot(rotten_tomato, regress_values, "b-")
plt.annotate(line_eq,(70,2500000000),fontsize=15,color="red")

#Create Labels
plt.title("Critical Reception (Tomatometer) vs Worldwide Box Office")
plt.xlabel("Tomatomater Score (Source: Rotten Tomatoes)")
plt.ylabel("Worldwide Box Office by USD Billions")

#Print Correlation
print(f'The correlation coefficient between our Rotten Tomatoes scores and the Worldwide Gross is {round(correlation[0],2)}')
plt.show()
```
